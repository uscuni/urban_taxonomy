```python
import os
import httpx
import pyogrio
import geopandas as gpd
import osmnx as ox

os.getenv("DOCKER_ENV_VERSION")
```

```python
url_vienna = "https://www.wien.gv.at/ma41datenviewer/downloads/Wien/FMZK/FMZK_WIEN_20240118_GP.zip"
r = httpx.get(url_vienna, verify=False)
```

```python
vienna = gpd.read_file(r.content)
vienna
```

```python
buildings_vienna = vienna[vienna["F_KLASSE_TEXT"].str.contains("Gebäude")]
buildings_vienna
```

```python
buildings_vienna.to_file("/data/buildings_raw/austria/vienna/vienna.gpkg")
```

```python
extensions = ["shp", "shx", "prj", "dbf", "qpj"]

for extension in extensions:
    url_linz = f"https://geo.data.linz.gv.at/katalog/geodata/stadtplan/gebaeudeflaechen/2018/gebaeudeflaechen_2018.{extension}"
    r = httpx.get(url_linz, verify=False)
    with open(f"/data/buildings_raw/austria/linz/linz.{extension}", "wb") as f:
            f.write(r.content)
```

```python
url_tirol = "https://data-tiris.opendata.arcgis.com/datasets/tiris::gebaeude-1.zip?outSR={%22wkid%22:31254}"
r = httpx.get(url_tirol, verify=False)
with open(f"/data/buildings_raw/austria/tirol/tirol.zip", "wb") as f:
    f.write(r.content)
```

```python
url_upper_austria = "https://e-gov.ooe.gv.at/at.gv.ooe.dorisdaten/DORIS_Basisdaten/GEBAEUDE_LOD1.zip"
r = httpx.get(url_upper_austria, verify=False)
with open(f"/data/buildings_raw/austria/upper_austria/upper_austria.zip", "wb") as f:
    f.write(r.content)
```

```python
tags = {"building": True}
states = ["Burgenland, Austria",
          "Carinthia, Austria",
          "Lower Austria, Austria",
          "Salzburg (state), Austria",
          "Styria, Austria",
          "Vorarlberg, Austria"]

for state in states:
    state_gdf = ox.features_from_place("Burgenland, Austria", tags)
    state_gdf = state_gdf[state_gdf.geometry.type == 'Polygon']
    state_gdf.to_file(f"/data/buildings_raw/austria/osm/{state}.gpkg")
```
