```python
import os

import geopandas as gpd
import tobler

from tqdm.auto import tqdm
```

```python
folder = "/data/uscuni-ulce/processed_data/clusters/"
morphotope_folder = "/data/uscuni-ulce/processed_data/morphotope_clusters/"
destination = "/data/uscuni-ulce/h3/"
```

```python
os.mkdir(destination)
os.mkdir(destination + "regions")
```

```python
regions = gpd.read_parquet("/data/uscuni-ulce/regions/cadastre_regions_hull.parquet")
```

```python
def generate_h3(region_id):
    morphotopes = gpd.read_parquet(f"{morphotope_folder}{region_id}_clusters.pq")
    h3_grid = tobler.util.h3fy(morphotopes.buffer(75), resolution=10)
    h3_interpolated = tobler.area_weighted.area_join(morphotopes, h3_grid, ['final', 'regional', 'final_without_noise'])
    h3_interpolated = h3_interpolated.dropna()
    h3_interpolated.to_parquet(f"{destination}{region_id}_clusters.pq")
```

```python
for region in tqdm(regions.index[::-1]):
    generate_h3(region)
```
