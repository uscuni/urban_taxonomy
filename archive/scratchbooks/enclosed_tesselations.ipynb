### remove overlapping building footprints

```python
overlapping = mm.CheckTessellationInput(buildings).overlap

i1, i2 = overlapping["geometry"].sindex.query(
    overlapping["geometry"], predicate="intersects"
)

overlapping_set = set()
for g1, g2 in zip(i1, i2, strict=False):
    if g1 == g2:
        continue
    overlapping_set.add(tuple(sorted((g1, g2))))
to_drop = [overlapping.iloc[g2].id for g1, g2 in overlapping_set]


buildings = buildings[
    ~np.isin(
        typed_dict.loc[buildings["id"].values].values, typed_dict.loc[to_drop].values
    )
]
```
